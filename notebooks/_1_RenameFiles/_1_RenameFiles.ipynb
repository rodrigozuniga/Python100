```python
import os
from datetime import datetime
import sys

dt = datetime.now()
dt = dt.date()
print(dt)

files = os.listdir('./data')

if 'a.txt' not in files:
    for file in files:
        new_name = file[0] + ".txt"
        os.rename(f'./data/{file}', f'./data/{new_name}')
else:
    files = os.listdir('./data')
    for file in files:
        print(file)
        new_name = file.replace('.txt',f"-{dt}.txt")
        os.rename(f'./data/{file}', f'./data/{new_name}')
        print(f"{file} renamed to {new_name}")

```
