```python
with open("snowwhite.txt") as f:
    string = f.read()
sentences = string.split('. ')
cap_sentences = [sentence.capitalize()+'. ' for sentence in sentences]
cap_string =" ".join(cap_sentences)
print(cap_string)
with open("snowwhite_corrected.txt", "w") as f:    
    f.write(cap_string)
```
